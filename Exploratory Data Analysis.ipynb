```python
import matplotlib.pyplot as plt 
import pandas as pd 
import seaborn as sns 
import numpy as np
```

```python
df = pd.read_csv('Titanic_Cleaned.csv')
```

```python
#Generate summary stats
df.describe()
```

```python
df[['PassengerId', 'Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].median()#median
```

```python
#Histogram for age
plt.figure(figsize=(8, 5))
plt.hist(df['Age'], bins=30, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#Histogram for Fare
plt.figure(figsize=(8, 5))
plt.hist(df['Fare'], bins=30, color='lightgreen', edgecolor='black')
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
#Boxplot for Age and Fare
plt.figure(figsize=(10, 5))

# Age Boxplot
plt.subplot(1, 2, 1)
sns.boxplot(y=df['Age'], color='orange')
plt.title('Age Boxplot')

# Fare Boxplot
plt.subplot(1, 2, 2)
sns.boxplot(y=df['Fare'], color='cyan')
plt.title('Fare Boxplot')

plt.tight_layout()
plt.show()
```

```python
# correlation matrix using only numeric columns
plt.figure(figsize=(10, 6))
numeric_df = df.select_dtypes(include=[np.number, 'bool'])
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
# To select only numeric columns from the dataframe:
numeric_df = df.select_dtypes(include=[np.number, 'bool'])
```
